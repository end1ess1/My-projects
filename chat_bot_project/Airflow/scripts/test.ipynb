```python
import time
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm

# Функция, которую будем выполнять параллельно
def task(n, overall_pbar):
    # Вложенный прогресс-бар для каждой задачи
    for i in tqdm(range(10), desc=f"Task {n}", position=n + 1, leave=False, dynamic_ncols=True):
        time.sleep(1)  # имитируем долгую задачу
    overall_pbar.update(1)  # Обновляем общий прогресс

# Количество задач
num_tasks = 10

# Инициализация прогресс-бара
with tqdm(total=num_tasks, desc="Overall Progress", position=0, dynamic_ncols=True) as overall_pbar:
    with ThreadPoolExecutor(max_workers=4) as executor:
        # Создаем задачи и отслеживаем их прогресс
        futures = []
        for i in range(num_tasks):
            future = executor.submit(task, i, overall_pbar)
            futures.append(future)
        
        # Ожидаем завершения всех задач
        for future in futures:
            future.result()

print("Все задачи завершены.")
```

```python
from concurrent.futures import ThreadPoolExecutor
from pymilvus import connections, FieldSchema, CollectionSchema, DataType, Collection

# Подключение к Milvus
connections.connect("default", host="localhost", port="19530")

# Определение схемы коллекции
fields = [
    FieldSchema(name="id", dtype=DataType.INT64, is_primary=True, auto_id=True),
    FieldSchema(name="embedding", dtype=DataType.FLOAT_VECTOR, dim=128)
]
schema = CollectionSchema(fields, "example_collection")
collection = Collection("example_collection", schema)

# Функция для загрузки одного документа
def load_document(doc):
    # Предположим, что doc — это вектор, который нужно загрузить
    entities = [
        [doc]  # Вектор документа
    ]
    collection.insert(entities)

# Список документов для загрузки (пример)
documents = [[float(i) for i in range(128)] for _ in range(1000)]

# Распараллеливание загрузки с использованием ThreadPoolExecutor
with ThreadPoolExecutor(max_workers=5) as executor:
    executor.map(load_document, documents)

# После загрузки можно создать индекс и загрузить коллекцию в память
collection.create_index("embedding", {"index_type": "IVF_FLAT", "metric_type": "L2", "params": {"nlist": 128}})
collection.load()

# Закрытие соединения
connections.disconnect("default")
```
