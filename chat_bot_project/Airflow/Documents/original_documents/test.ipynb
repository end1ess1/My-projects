```python
def split_sliding_window(text, chunk_size=512, overlap=128):
    words = text.split()
    chunks = []
    for i in range(0, len(words), chunk_size - overlap):
        chunks.append(" ".join(words[i:i+chunk_size]))
    return chunks
```

```python
with open('main.txt', 'r', encoding='utf-8') as ff:
    data = ff.read()
```

```python
data
```

```python
splitted_data = split_sliding_window(data)
```

```python
import tiktoken

# Загружаем токенизатор для GPT
encoding = tiktoken.encoding_for_model("gpt-3.5-turbo")

# Пример предложения
sentence = "Правила приема на 2025/2026 учебный год вступают в силу с 1 марта 2025 года!"

# Токенизируем предложение
tokens = encoding.encode(sentence)

print("ID токенов:", tokens)
print("Количество токенов:", len(tokens))
```
